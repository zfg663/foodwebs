## Script to plot spectra

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
data = np.loadtxt('data/random_uniform/random_eigenvalues_N02.txt') # load

real = np.concatenate([data[:,0], data[:,2]])
imag = np.concatenate([data[:,1], data[:,3]])

fig, ax = plt.subplots(2,3, figsize=(10,4), sharex=True, sharey=True)
ax[0,0].hist2d(real, imag, bins=[100,50], norm = mpl.colors.LogNorm());
fig;
```
