```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# PARAMETERS

# font in plot
plt.rc('font', **{'family': 'serif', 'serif': ['stixgeneral']})

# file names of in and out put
folder = 'data/2/'
filename = 'from_data_2_3'

# iterations to be plotted
fw_i = [8099, 8129, 9131, 9137, 9470, 9864] 
#fw_i = [1537 + i for i in range()]

# number off plots
n_webs = len(fw_i)
```

```python
# empty arrays for importing data
data = [[] for i in range(n_webs)]
steady_data = [[] for i in range(n_webs)] 

# IMPORTING DATA
# importing time series
for i in range(n_webs):
    iter = str(fw_i[i])
    data[i] = np.loadtxt('data/timeseries_' + iter + '.txt')
    #steady_data = np.loadtxt('C:/Users/Bruker/Documents/Uni/Bachelor project/code/code/data/plot/' + direc +'steady_states_'+iter+'.txt')

# importing food web characteristics
basal = np.loadtxt('data/basal_par.txt')
species = np.loadtxt('data/species_par.txt')
stab = np.loadtxt('data/stab_data.txt')
```

```python
#%% INITIALIZING ARRAYS

# empty arrays for data
S = [[] for m in range(n_webs)]
t = [[0] for m in range(n_webs)]
n = [[] for m in range(n_webs)]

# arrays for visuals in plot
labels = [[] for m in range(n_webs)]
colors = [[] for m in range(n_webs)]

for fw in range(n_webs):
    # identifying indices in files of chosen iteration
    i = 0
    while stab[i,0] < fw_i[fw]:
        i += 1
     
    # number of species    
    n[fw] = int(stab[i,1])
        
    # index array for keeping track of extinct species
    index = [m for m in range(n[fw])]
    ext_index = []
        
    # steady states
    #for i in range(0,n):
    #    steady_states[i] = steady_data[i+1]
    #steady_states = [0.597716, 0.195022, 0.0850549, 0.184205, 0.12999, 0.0516673, 0.0768291]
    
    # density array
    S[fw] = [[data[fw][0,2*m+1]] for m in range(n[fw])];
    colors[fw] = ['cornflowerblue' for m in range(n[fw])]
    for m in range(n[fw]):
        if S[fw][m][0] < 1e-9:
            colors[fw][m] = 'orange'

#   SORTING DATA IN ARRAYS
    t_0 = 0;
    
    for m in range (len(data[fw])-1):
        t[fw] += [data[fw][m,0] + t_0]
        k = 0
        for j in index:
            S[fw][j] += [data[fw][m,2*k+1]]
            k += 1
#        for j in ext_index:
#            S[fw][j] += [0]
            
        # for zooming in    
        # if data[i+1,0] == 0:
        #    break
        
        # if extinction
        if data[fw][m+1,0] == 0:
            print(m)
            for j in index:
                if S[fw][j][m+1] <= 2e-12:
                    colors[fw][j] = 'r'
                    t_0 = t[fw][m+1]
                    ext_index.append(j)
                    if j == index[len(index)-1]:
                        index.remove(j)
                    else:
                        ii = 0
                        for idx in index:
                            if idx == j:
                                index[ii] = index[len(index)-1]
                            ii += 1
                        del index[len(index)-1]
                  
              
    # labels
    labels[fw] = ['' for m in range(0,n[fw])]
    b_min = 0
    while basal[b_min,0] < fw_i[fw]:
        b_min = b_min + 1
        
    b_max = b_min
    while basal[b_max,0] == fw_i[fw]:
        #labels[fw][b_max-b_min] =  'Species ' + str(int(basal[b_max, 1])) + ', (producer)'
        b_max = b_max + 1
    #if basal[b_max - 1,4] == 0.01:
    #    b_max -= 1  
    
    s_min = 0
    while species[s_min,0] < fw_i[fw]:
        s_min = s_min + 1
       
    s_max = s_min
    while species[s_max,0] == fw_i[fw]:
        #labels[fw][b_max-b_min + s_max-s_min] = 'Species ' + str(int(species[s_max, 1]))
        s_max = s_max + 1
```

```python
# PLOTTING

# creating figure    
fig = plt.figure()
fig, ax = plt.subplots(1, n_webs, sharex=False, sharey=True,figsize=(24,5))

# axis configurations
fig.text(0.53, -0.1, 'Time (log)', ha='center', fontsize=40)
ax[0].set_ylabel('Density', fontsize=40)
ax[0].set_yticks([0, 0.5, 1])


# Looping through plots
for fw in range(n_webs):
    
#   Looping through species in plot
    for i in range(n[fw]):
        ax[fw].plot(t[fw][0:len(S[fw][i])], S[fw][i], color = colors[fw][i])
        
#   Adding steady states    
    #for i in range(0,n):
    #   line = np.array([steady_states[i] for k in range(len(t))])
    #   plt.plot(t, line, '--', color = 'lightgrey')

#   Ticks
    ax[fw].tick_params(axis = "both", labelsize=40)
    #ax[fw].set_xticks([])

#   Changing scale of plots
    #ax[fw].set_xlim(1, 10000)
    #ax[fw].set_xscale('log')
    #ax[fw].set_yscale('log')
    

# Title and legends
#ax[2].legend(labels[2], loc = 'upper right', prop={'size': 20})
#plt.title('Time Series for Iteration ' + iter, fontsize=26)
```
